implementing PCA manually

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
```

```python
print(X,y)
```

```python
import numpy as np

# X shape: (n_samples, 4_features)

mean = np.mean(X, axis=0)   # shape (4,)
X_c = X - mean
```

```python
cov_X=(X_c.T @ X_c) / (X.shape[0]-1)
```

```python
eigvals,eigvecs=np.linalg.eigh(cov_X)
print("Eigenvalues:", eigvals)
print("Eigenvectors:\n", eigvecs)
```

```python
idx = np.argsort(eigvals)[::-1]
eigvals_sorted = eigvals[idx]
eigvecs_sorted = eigvecs[:, idx]
print(eigvals_sorted)
print(eigvecs_sorted)
```

```python
W = eigvecs_sorted[:, :2]
print(W)
```

```python
Z = X_c @ W
print(Z)
```

```python
print("Explained variance ratio:", eigvals_sorted / np.sum(eigvals_sorted))
```

```python
import matplotlib.pyplot as plt

plt.scatter(Z[:,0], Z[:,1], c=y)
plt.title("PCA on Iris")
plt.show()
```
